#### Task 14
Study the code below, which displays player name, date, minutes, points, and points per minute for the ten best points-per-minute games. It only counts games where the player played for 25 minutes or more. 

```python
eligible_games = nba.loc[nba['MIN'] >= 25, ['Player', 'Date', 'MIN', 'PTS', 'PTS_PM']]
eligible_games[eligible_games['PTS_PM'].rank(method = 'dense', ascending = False) <= 10]
```

### Task 15
Show player name, total steals, total turnovers, and season steal to turnover ratio for players with at least 30 steals. Display in decreasing order of steal to turnover ratio and show the first 10 rows only. Hint: You can break this into separate steps.

```python
stl_to = nba.groupby('Player').agg(STL = ('STL', 'sum'),
                                   TO = ('TO', 'sum'))
stl_to['ST_Ratio'] = stl_to['STL'] / stl_to['TO']

qual_stl_to = stl_to.loc[stl_to['STL'] >= 30, ['STL', 'TO', 'ST_Ratio']]

qual_stl_to.sort_values('ST_Ratio', ascending=False).head(10)
```

#### Task 16
Show the data sorted in descending order of scoring average (average points). Show only the player, team, and average points. Try to show the result as a DataFrame type, rather than as a Series. Show the first 25 rows.

```python
avg_pts = nba.groupby('Player').agg({'Team': 'unique', 'PTS': 'mean'})

avg_pts.sort_values('PTS', ascending=False).head(25)
```

#### Task 17
Show the data sorted in descending order of points scored per minute for players who have played at least 1,000 minutes. Show only the player, team, and points per minute. Try to show the result as a DataFrame type, rather than as a Series. Show the first 25 rows. Note that you need to calculate points per minute as total points divided by total minutes, not as the mean of the points per minute column.

```python
qual_avg_pts = nba.groupby('Player').agg({'Team': 'unique', 'MIN': 'sum', 'PTS': 'sum'})

qual_avg_pts['PTS_MIN'] = qual_avg_pts['PTS'] / qual_avg_pts['MIN']

qual_avg_pts = qual_avg_pts.loc[qual_avg_pts['MIN'] >= 1000, ['Team', 'PTS_MIN']]

qual_avg_pts.sort_values('PTS_MIN', ascending=False).head(25)
```

#### Task 18
Show the top 30 rows of player name and total minutes played, ordered in descending order of total minutes played.
