# Scatter plot example

```python
df=pd.read_csv('nba_players.csv')
```

```python
player_age= df['AGE ']
player_min= df['MIN ']
```

```python
fig,ax=plt.subplots()
ax.scatter( player_age, player_min)
ax.set_xlabel('Age')
ax.set_ylabel('Min')
plt.show()
```

### Top 5 players and their data

```python
df_sorted=df.sort_values(by=['PTS '],ascending= False)

player= df_sorted.head()['PLAYER ']

points= df_sorted.head()['PTS ']
```

```python
plt.rcdefaults()
fig, ax = plt.subplots()

ax.barh(range(len(player)), points, align='center')

ax.set_yticks((range(len(player))), player) 

ax.invert_yaxis()  
ax.set_xlabel('Points')
ax.set_title('Player with Highest Points')

plt.show()
```
